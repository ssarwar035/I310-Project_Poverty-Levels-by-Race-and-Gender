#**SVR Model Training and Scaling + Data Visualization**


```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import math

# Load the initial data
df_initial = pd.read_csv("poverty_level_wages.csv")

# Separate target from predictors
X_initial = df_initial.drop(['year', 'annual_poverty-level_wage'], axis=1)
y_initial = df_initial['annual_poverty-level_wage']

# Scale data
scaler = MinMaxScaler()
X_scaled_initial = scaler.fit_transform(X_initial)

# Build SVR model on the initial data
svr = SVR()
svr.fit(X_scaled_initial, y_initial)

# Load the future data (2015 to 2020)
future_1 = pd.read_csv('future_data.csv')

# Check and print column names to ensure consistency
print("Column Names in future_1:", future_data.columns)

# Ensure that the features used during training match the features in future_1
X_future = future_data.drop(['year'], axis=1)  # Assuming 'year' is the only column to drop
X_future_scaled = scaler.transform(X_future)

# Use the trained SVR model to make predictions for the future data
future_predictions = svr.predict(X_future_scaled)

# Load actual values for the future data from a different CSV file ('poverty_level_wages.csv')
df_future_actual = pd.read_csv('poverty_level_wages.csv')
y_future_actual = df_future_actual[df_future_actual['year'].between(2015, 2020)]['annual_poverty-level_wage']

# Calculate RMSE and R-squared
rmse = math.sqrt(mean_squared_error(y_future_actual, future_predictions))
r_squared = r2_score(y_future_actual, future_predictions)

# Scatter plot of Actual and Predicted Values for the years 2015 to 2020
plt.figure(figsize=(12, 8))
plt.scatter(future_1['year'], y_future_actual, label='Actual Values', color='blue')
plt.scatter(future_1['year'], future_predictions, label='Predicted Values', color='red', marker='x')
plt.title(f'Actual vs Predicted Values Over Time (2015 to 2020)\nRMSE: {rmse:.4f}, R-squared: {r_squared:.4f}')
plt.xlabel('Year')
plt.ylabel('Wages Below Poverty Level')
plt.legend()
plt.show()

print("Root Mean Squared Error (RMSE):", rmse)
print("R-squared:", r_squared)
```

# **Predictive Extension to the next 30 years**

**The only part that needs to be changed is this, added a loop structure for predicting the next 30 years **

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

df = pd.read_csv('poverty_level_wages.csv')
X = df['year'].values.reshape(-1,1)
y = df['annual_poverty-level_wage'].values

model = LinearRegression()
model.fit(X, y)

start_year = 2020
num_years = 31
X_future = [[year] for year in range(start_year, start_year + num_years)]
predictions = model.predict(X_future)

plt.scatter(df['year'], y, color='black')
plt.plot(df['year'], model.predict(X), color='blue')
future_years = range(start_year, start_year + num_years)
plt.plot(future_years, predictions, color='red')
plt.title('Annual Wages Below Poverty Level')
plt.xlabel('Year')
plt.ylabel('Wages Below Poverty Level')
plt.legend(['Historical', 'Linear Model', 'Predictions'])
plt.grid()

plt.show()
```
