```python
train_data = df[(df['year'] >= 1973) & (df['year'] <= 2000)]
test_data = df[(df['year'] >= 2015) & (df['year'] <= 2020)]

X_train = train_data.drop('annual_poverty-level_wage', axis=1)
y_train = train_data['annual_poverty-level_wage']

X_test = test_data.drop('annual_poverty-level_wage', axis=1)
y_test = test_data['annual_poverty-level_wage']
```

```python
model = RandomForestRegressor()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f'Root Mean Squared Error: {rmse}')
```
